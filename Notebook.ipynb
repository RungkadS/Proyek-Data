```python
import os
import pandas as pd

def merge_csv_files(folder_path, output_file):
  all_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
  df_list = []

  for filename in all_files:
    file_path = os.path.join(folder_path, filename)
    df = pd.read_csv(file_path)
    df_list.append(df)

  merged_df = pd.concat(df_list, ignore_index=True)
  merged_df.to_csv(output_file, index=False)

folder_path = '/content/drive/MyDrive/Project_Data/data'
output_file = '/content/drive/MyDrive/Project_Data/data/all.csv'

merge_csv_files(folder_path, output_file)
```

Cleaning missing value pada setiap csv file menggunakan metode interpolasi karena cocok dengan data Timeseries

```python
import glob
import os

def interpolate_timeseries(df):
    # Melakukan interpolasi untuk cleaning missing value pada semua csv file
    df['PM2.5'].interpolate(method='linear', limit_direction='forward', inplace=True)
    df['PM10'].interpolate(method='linear', limit_direction='forward', inplace=True)
    df.SO2.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.NO2.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.CO.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.O3.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.TEMP.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.PRES.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.DEWP.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.RAIN.interpolate(method='linear', limit_direction='forward', inplace=True)
    df.WSPM.interpolate(method='linear', limit_direction='forward', inplace=True)
    return df

def process_all_csv_files(directory_path):
    all_csv_files = glob.glob(os.path.join(directory_path, "*.csv"))

    for file in all_csv_files:
        data = pd.read_csv(file)

        data_interpolated = interpolate_timeseries(data)

        data_interpolated.to_csv(file, index=False)
        print(f"Interpolasi selesai untuk file: {file}")

directory_path = '/content/drive/MyDrive/Project_Data/data'
process_all_csv_files(directory_path)
```

```python
#raplace nilai max untuk PM2.5 dan PM10 dengan nilai max masing-masing 500 dengan konversi nilai konsentrasi Indeks Kualitas Udara
def replace_values(df):
# Untuk nilai diatas 500 nilainya akan dikali dengan 0.1 karena nilai maksimal konsentrasi polutan PM2.5 dan PM10 diangka 500
    df['PM2.5'] = np.where(df['PM2.5'] > 500, df['PM2.5'] * 0.1, df['PM2.5'])
    df['PM10'] = np.where(df['PM10'] > 500, df['PM10'] * 0.1, df['PM10'])

replace_values(aotizhongxin_df)
```

```python
# Mengonversi 'date'
aotizhongxin_df['date'] = pd.to_datetime(aotizhongxin_df[['year', 'month', 'day']])

# Menentukan tanggal terakhir dalam data
last_date = aotizhongxin_df['date'].max()
# Filter untuk data 1 bulan trakhir
one_month_ago = last_date - pd.DateOffset(months=1)

# Filter data 1 bulan terakhir
last_month_data = aotizhongxin_df[aotizhongxin_df['date'] >= one_month_ago]

# Groupby berdasarkan tanggal harian dan memunculkan rata-rata harian temperature
daily_avg_TEMP_aotizhongxin = last_month_data.groupby(last_month_data['date'].dt.date).agg({
    "TEMP": ("mean", "max", "min")

}).reset_index()
print(daily_avg_TEMP_aotizhongxin)
```

```python
# Mengonversi 'date' jika dataset belum memiliki kolom bertipe datetime untuk year dan month
aotizhongxin_df['bulan'] = pd.to_datetime(aotizhongxin_df[['year', 'month']].assign(day=1))  # Set tanggal ke 1 untuk representasi bulanan

# Menentukan bulan terakhir dalam data
last_month = aotizhongxin_df['bulan'].max()

# Hitung satu tahun yang lalu dari bulan terakhir
one_year_ago = last_month - pd.DateOffset(years=1)

# Filter data untuk satu tahun terakhir
last_year_data = aotizhongxin_df[aotizhongxin_df['bulan'] >= one_year_ago]

# Groupby berdasarkan bulan dan menghitung rata-rata untuk setiap polutan per bulan
monthly_avg_pollutant_aotizhongxin = last_year_data.groupby(last_year_data['bulan'].dt.to_period('M')).agg({
    "PM2.5": "mean",
    "PM10": "mean",
    "SO2": "mean",
    "NO2": "mean",
    "CO": "mean",
    "O3": "mean"
}).reset_index()

# Menampilkan hasil
print(monthly_avg_pollutant_aotizhongxin)
```

```python
# Mengkonversikan nilai PM2.5 menjadi AQI(Air Quality Index)
bins = [0, 50, 100, 150, 200, 300, float('inf')]
kategori = ['Baik', 'Sedang', 'Tidak sehat untuk orang sensitif', 'Tidak Sehat', 'Sangat Tidak Sehat', 'Berbahaya']

monthly_avg_AQI_aotizhongxin = monthly_avg_pollutant_aotizhongxin.copy()
monthly_avg_AQI_aotizhongxin['PM2,5'] = monthly_avg_pollutant_aotizhongxin['PM2.5']

# rumus menghitung AQI konversi dari PM2.5
# https://www.iqair.com/id/newsroom/what-is-aqi?srsltid=AfmBOop1gD7-F6Ac9SI-EJKB34xhI1Hx1RAml2hC641IbuQRdxCuXtcn
def calculate_ispu(pm25):
    if 0 <= pm25 <= 35:
        return 0 + ((50 - 50) / (35 - 0)) * (pm25 - 0)
    elif 35.1 <= pm25 <= 75:
        return 51 + ((100 - 51) / (75 - 35.1)) * (pm25 - 35.1)
    elif 75.1 <= pm25 <= 115:
        return 101 + ((150 - 101) / (115 - 75.1)) * (pm25 - 75.1 )
    elif 115.1 <= pm25 <= 150:
        return 151 + ((200 - 151) / (150 - 115.1)) * (pm25 - 115.1)
    elif 150.1 <= pm25 <= 250:
        return 201 + ((300 - 201) / (250 - 150.1)) * (pm25 - 150.1)
    else:
        return 301  # AQI di atas 300 dianggap 'Berbahaya'

# Menambahkan kolom baru AQI berdasarkan rata-rata PM2.5 per bulan
monthly_avg_AQI_aotizhongxin['AQI'] = monthly_avg_AQI_aotizhongxin['PM2,5'].apply(calculate_ispu)
# menambahkan kolom baru kategori ISPU berdasarkan rata-rata PM2.5 per bulan
monthly_avg_AQI_aotizhongxin['Kategori_AQI'] = pd.cut(monthly_avg_AQI_aotizhongxin['AQI'], bins=bins, labels=kategori, right=False, include_lowest=True, duplicates='drop')
# Menampilkan hasil AQI-nya perbulan dalam 1 tahun terakhir
print(monthly_avg_AQI_aotizhongxin[['bulan', 'PM2.5', 'AQI', 'Kategori_AQI']])
```
